```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import TimeSeriesSplit
import time

results_Sparse = []
for i in range(1,15):
    print(i)
    dataset = pd.read_csv(f"JL_Sparse/JL_Sparse_{i}.csv")

    X = dataset.drop(columns=['Close'])
    y = dataset['Close']
    
    timesplit= TimeSeriesSplit(n_splits=10)
    for train_index, test_index in timesplit.split(X):
        X_train, X_test = X[:len(train_index)], X[len(train_index): (len(train_index)+len(test_index))]
        y_train, y_test = y[:len(train_index)].values.ravel(), y[len(train_index): (len(train_index)+len(test_index))].values.ravel()


    model = LinearRegression()
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_r2 = r2_score(y_train, y_train_pred)

    print(f"Training MSE: {train_mse:.3f}")
    print(f"Training MAE: {train_mae:.3f}")
    print(f"Training R²: {train_r2:.3f}")

    y_test_pred = model.predict(X_test)

    test_mse = mean_squared_error(y_test, y_test_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)

    print(f"Test MSE: {test_mse:.3f}")
    print(f"Test MAE: {test_mae:.3f}")
    print(f"Test R²: {test_r2:.3f}")

    results_Sparse.append([train_mse, train_mae, train_r2, test_mse, test_mae, test_r2])
```

```python
import time 

start_time = time.time()
dataset = pd.read_csv(f"Hello.csv")

X = dataset.drop(columns=['Date','Close'])
y = dataset['Close']

timesplit= TimeSeriesSplit(n_splits=10)
for train_index, test_index in timesplit.split(X):
    X_train, X_test = X[:len(train_index)], X[len(train_index): (len(train_index)+len(test_index))]
    y_train, y_test = y[:len(train_index)].values.ravel(), y[len(train_index): (len(train_index)+len(test_index))].values.ravel()


model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
train_mse = mean_squared_error(y_train, y_train_pred)
train_mae = mean_absolute_error(y_train, y_train_pred)
train_r2 = r2_score(y_train, y_train_pred)

print(f"Training MSE: {train_mse:.3f}")
print(f"Training MAE: {train_mae:.3f}")
print(f"Training R²: {train_r2:.3f}")

y_test_pred = model.predict(X_test)

test_mse = mean_squared_error(y_test, y_test_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)

print(f"Test MSE: {test_mse:.3f}")
print(f"Test MAE: {test_mae:.3f}")
print(f"Test R²: {test_r2:.3f}")

print("--- %s seconds ---" % (time.time() - start_time))
```
